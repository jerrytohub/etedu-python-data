# 전국 초중고등학교 위치 산점도 그리기

```python
# 데이터 읽기 및 준비하기
import csv
f = open('school.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
```

```python
# 위도, 경도, 색깔 데이터를 리스트에 넣기
WD = []
GD = []
colors = []

for row in data:
    row[2:] = map(float, row[2:])
    WD.append(row[2])
    GD.append(row[3])
    if row[1] == '초등학교':
        colors.append('red')
    elif row[1] == '중학교':
        colors.append('blue')
    else:
        colors.append('green')
```

```python
# 산점도 그리기
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.scatter(GD, WD, s=1, color=colors)
plt.xlabel('경도')
plt.ylabel('위도')
plt.title('전국 초중고등학교 위치')
plt.show()
```

# 가장 가까운 학교를 알려주는 프로그램 만들기

```python
# 데이터 읽기 및 준비하기
import csv
f = open('school.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
```

```python
# 학교급과 위치 입력받기
grade = input('초등학교, 중학교, 고등학교 중 골라주세요.')
WD = float(input('당신이 사는 지역의 위도는 무엇입니까?'))
GD = float(input('당신이 사는 지역의 경도는 무엇입니까?'))
```

```python
# 변수 설정하기
min = 1000000
school = ''
```

```python
# 최단 거리의 최솟값 찾기
for row in data:
    row[2:] = map(float, row[2:])
    if (grade == row[1]):
        distance = (GD-row[3])*(GD-row[3]) + (WD-row[2])*(WD-row[2])
        if distance < min:
            min = distance
            school = row[0]
```

```python
# 최단 거리가 최솟값일 때의 학교 이름 출력하기
print('당신의 집과 가장 가까운 학교는 ' + school + '입니다.')
```

```python
# 완성코드
import csv
f = open('school.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

grade = input('초등학교, 중학교, 고등학교 중 골라주세요.')
WD = float(input('당신이 사는 지역의 위도는 무엇입니까?'))
GD = float(input('당신이 사는 지역의 경도는 무엇입니까?'))

min = 1000000
school = ''

for row in data:
    row[2:] = map(float, row[2:])
    if (grade == row[1]):
        distance = (GD-row[3])*(GD-row[3]) + (WD-row[2])*(WD-row[2])
        if distance < min:
            min = distance
            school = row[0]
                         
print('당신의 집과 가장 가까운 학교는 ' + school + '입니다.')
```
