```python
# 데이터 읽기 및 준비하기
import csv
f = open('price.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
```

```python
# 사용자에게 간식 입력받기
food = input('<품목>'
             '\n빵/사탕/아이스크림/스낵과자/주스/탄산음료/떡볶이/치킨/햄버거/피자' 
             '\n어떤 간식의 가격 변화를 보고 싶은가요?')
```

```python
# 간식의 연도별 물가 데이터를 리스트에 넣기
price = []

for row in data:
    if food  == row[0]:
        for i in row[1:]:
            price.append(float(i))
```

```python
# 연도 데이터 리스트 만들기
years = []
for i in range(2003, 2024):
    years.append(i)
```

```python
# 선 그래프 그리기
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.plot(years, price, label = food)
plt.xlabel('연도')
plt.ylabel('물가지수')
plt.title(food + '의 물가 변화')
plt.legend()
plt.show()
```

```python
# 완성 코드
import csv
f = open('price.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

food = input('<품목>'
             '\n빵/사탕/아이스크림/스낵과자/주스/탄산음료/떡볶이/치킨/햄버거/피자' 
             '\n어떤 간식의 가격 변화를 보고 싶은가요?')

price = []
for row in data:
    if food  == row[0]:
        for i in row[1:]:
            price.append(float(i))
            
years = []
for i in range(2003, 2024):
    years.append(i)
    
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.plot(years, price, label = food)
plt.xlabel('연도')
plt.ylabel('물가지수')
plt.title(food + '의 물가 변화')
plt.legend()
plt.show()
```
