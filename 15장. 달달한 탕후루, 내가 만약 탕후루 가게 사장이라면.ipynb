# 서울시 지역의 연령대별 인구 비율 원그래프

```python
# 판다스로 데이터 읽기 및 준비하기
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('202310인구.csv', encoding='cp949', index_col=0)
```

```python
# '50세 이상' 열 생성하기
df['50세 이상'] = df.iloc[:, 7:].sum(axis=1)
df.head(2)
```

```python
# 필요 없는 열 삭제
df = df.drop(columns=df.iloc[:, 7:13])
df.head(2)
```

```python
# 필요한 데이터만 선택하여 저장하기
area = df.index
label = df.columns[3:]
print(area)
print(label)
```

```python
# 원그래프로 나타내기
for i in range(0, 25):
    plt.rc('font',family='Malgun Gothic')
    plt.pie(df.iloc[i, 3:], labels=label, autopct='%.1f%%')
    plt.title(area[i]+' 지역의 연령대별 인구 비율')
    plt.axis('equal')
    #plt.savefig(area[i]+' 지역의 연령대별 인구 비율'+'.png')  # 그림 파일을 저장하고 싶다면 #을 삭제해요.
    plt.show()
```

```python
#완성코드
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('202310인구.csv', encoding='cp949', index_col=0)

df['50세 이상'] = df.iloc[:, 7:].sum(axis=1)
df = df.drop(columns=df.iloc[:, 7:13])

area = df.index
label = df.columns[3:]

for i in range(0, 25):
    plt.rc('font',family='Malgun Gothic')
    plt.pie(df.iloc[i, 3:], labels=label, autopct='%.1f%%')
    plt.title(area[i]+' 지역의 연령대별 인구 비율')
    plt.axis('equal')
    #plt.savefig(area[i]+' 지역의 연령대별 인구 비율'+'.png')  # 그림 파일을 저장하고 싶다면 #을 삭제해요.
    plt.show()
```

# 원그래프의 explode 속성

```python
#explode 속성 지정하기
explode = [0, 0.1, 0, 0, 0]
for i in range(0, 25):
    plt.rc('font',family='Malgun Gothic')
    plt.pie(df.iloc[i, 3:], labels=label, autopct='%.1f%%', explode=explode)
    plt.title(area[i]+' 지역의 연령대별 인구 비율')
    plt.axis('equal')
    #plt.savefig(area[0]+' 지역의 연령대별 인구 비율'+'.png')
    plt.show()
```
